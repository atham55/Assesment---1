```python
# matrimony
import csv
import os
import logging
#filename='mohamedatham.csv'
class matrimony:
    filename='ma.csv'
    def insertion(self):
        self.uniqueid=int(input('enter uniqueid :'))
        self.name=input('enter name :')
        self.gender=input('enter gender :')
        self.age=int(input('enter age :'))
        self.occupation=input('enter occupation :')
        self.salary=int(input('enter salaray :'))
        self.likes=input('enter likes :')
        self.dislikes=input('enter dislikes :')
        self.location=input('enter location :')
        self.degree=input('enter the degree :')

        with open(self.filename,'a',newline='')as x:
            csvwr=csv.writer(x)
            csvwr.writerow([self.uniqueid,self.name,self.gender,self.age,self.occupation,self.salary,self.likes,self.dislikes,self.location,self.degree])
    def retrival(self):
        print('*'*34)
        print('uniqueid\tname\tgender\tage\toccupation\tsalary\tlikes\tdislikes\tlocation\tdegree')
        print('*'*34)
        with open(self.filename,'r')as x:
            d=csv.reader(x)
            for i in list(d):
                print(f'{i[0]}\t\t{i[1]}\t\t{i[2]}\t\t{i[3]}\t\t{i[4]}\t\t{i[5]}\t\t{i[6]}\t\t{i[7]}\t\t{i[8]}\t\t{i[9]}')
                print('-'*34)
        print('your data is successfully')
    def searching(self):
        se=input('enter the name to search ')
        print('uniqueid\tname\tgender\tage\toccupation\tsalary\tlikes\tdislikes\tlocation\tdegree')
        with open(self.filename,'r')as x:
            d=csv.reader(x)
            namelist=list(d)
            result=[names for names in namelist if se in names]
            if len(result)>0:
                for i in result:
                    print(f'{i[0]}\t\t{i[1]}\t\t{i[2]}\t\t{i[3]}\t\t{i[4]}\t\t{i[5]}\t\t{i[6]}\t\t{i[7]}\t\t{i[8]}\t\t{i[9]}')
            else:
                print('no result found')
    def deletion(self):
        dele=input('enter the person name to delete')
        with open(self.filename,'r')as x:
            d=csv.reader(x)
            namelist=list(d)
            f_result=[names for names in namelist if dele !=names[0]]
        with open(self.filename,'w',newline='')as x:
            csvwr=csv.writer(x)
            csvwr.writerows(f_result)
            if(len(f_result)==len(namelist)):
                print('nothing is available to delete')
            else:
                print('updated list')
        self.retrival()
        
    def menu(self):
        print('menu')
        print('1:insertion')
        print('2:retrival')
        print('3:searching')
        print('4:deletion')
        c=None
        try:
            c=int(input('enter the choice:'))
        except:
            logging.error('enter the choice:')
        return c
        print('welcome to matrimony application')
db=matrimony()
c=db.menu()
while True:
    if c==0:
        break
    elif c==1:
        db.insertion()
        break
    elif c==2:
        db.retrival()
        break
    elif c==3:
        db.searching()
        break
    elif c==4:
        db.deletion()
        break
    else:
        pass
```
